```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
g=9.81
vtb=6.8
vttmax=100/3.6

```

```python
N=1000
tfinal=5
yt= np.empty(N)
vt= np.empty(N)
t= np.empty(N)
yt[0]=0
vt[0]=0
t[0]=0
ytt= np.empty(N)
vtt= np.empty(N)
tt= np.empty(N)
ytt[0]=0
vtt[0]=0
tt[0]=0
dt=tfinal/N
```

```python
def euler(N, yt, vt, t, dt, vmax):
    zerot=0
    for i in range(N-1): #ERRADOOOOOO, TEM QUE TER A EXPRESSAO DA ACELERAÇÃO COM DRAG, E AÍ A VMAX FICA CONTEMPLADA NA EXPRESSAO NAO É PRECISO IF ELSE
        a=g
        yt[i+1]=yt[i]+vt[i]*dt
        if (vt[i]<vmax):
            vt[i+1]=vt[i]+a*dt
        else: 
            vt[i+1]=vmax
        t[i+1]=t[i]+dt
        if (yt[i+1]>5 and yt[i]<5):
            zerot=((2*i+1)/2)*dt
    return zerot
badt=euler(N, yt, vt, t, dt, vtb)
plt.plot(t,yt, linestyle="-")
tenist=euler(N, ytt, vtt, tt, dt, vttmax)
plt.plot(tt,ytt, linestyle="-")
plt.show()
```
