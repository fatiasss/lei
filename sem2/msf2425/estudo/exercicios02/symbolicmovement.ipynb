```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
t, yt, vt, at, atex = sy.symbols("t, yt, vt, at, atex")
g = 9.81
vm = 6.80
t_vals = np.linspace(0, 4, 1000)
yt = ((vm**2)/g)*sy.log(sy.cosh((g*t)/vm))
yt_lam = sy.lambdify(t, yt, "numpy")
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t_vals, yt_lam(t_vals), linestyle="-") #-yt para inverter as coordenadas y
```

```python
vt = sy.diff(yt, t)
vt_lam = sy.lambdify(t, vt, "numpy")
axis[0,1].plot(t_vals, vt_lam(t_vals), linestyle="-")
figure
```

```python
at = sy.diff(vt, t)
at_lam = sy.lambdify(t, at, "numpy")
axis[1,0].plot(t_vals, at_lam(t_vals), linestyle="-")
figure
```

```python
atex = g - (g/(vm**2))*abs(vt)*vt
atex_lam = sy.lambdify(t, atex, "numpy")
axis[1,0].plot(t_vals, atex_lam(t_vals), linestyle="-")
figure
```
