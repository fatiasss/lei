```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
g=9.8
```

```python
N=10000
tfinal=4
yt= np.empty(N)
vt= np.empty(N)
t= np.empty(N)
yt[0]=0
vt[0]=0
t[0]=0
dt=tfinal/N
```

```python
def euler(N, yt, vt, t, dt):
    for i in range(N-1):
        a=-g
        yt[i+1]=yt[i]+vt[i]*dt
        vt[i+1]=vt[i]+a*dt
        t[i+1]=t[i]+dt
    return
euler(N, yt, vt, t, dt)
```

```python
figure, axis= plt.subplots(2,2)
axis[0,0].plot(t,yt, linestyle="-")
axis[0,1].plot(t,vt, linestyle="-")
vt[int(3/dt)] #quanto maior o N, mais se aproxima de -29.4, o resultado real; -29.4 = -g*t, t=3
```

```python
yt[int(2/dt)] #posição aos 2 segundos em metros, de novo quanto maior o N, mais se aproxima de (-g/2*)2**2=-19.6
```

```python
time, xt = sy.symbols("time, xt")
xt = (-g/2)*time**2
xt_lam= sy.lambdify(time, xt, "numpy")
trueval = xt_lam(2)
trueval
```

```python
maxN= 1000
desvios = np.zeros(maxN)
dtvals = np.zeros(maxN)
for i in range(200, maxN-1): #calculo dos desvios em função de N
    yt= np.empty(i)
    vt= np.empty(i)
    t= np.empty(i)
    yt[0]=0
    vt[0]=0
    t[0]=0
    dt=tfinal/i
    euler(i, yt, vt, t, dt)
    twoval = yt[int(2/dt)]
    desvios[i] = abs(trueval-twoval)
    dtvals[i]=dt
axis[1,0].scatter(dtvals, desvios)
figure
```
