```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
vmaxb=-6.8
vmaxt=-100/3.6
g=-9.81
y0= 5
```

```python
t, ytb, ytt = sy.symbols("t, ytb, ytt")
ytb= ((vmaxb**2)/g)*sy.log(sy.cosh((g*t)/vmaxb)) + 5
ytt= ((vmaxt**2)/g)*sy.log(sy.cosh((g*t)/vmaxt)) + 5
ytb_lam= sy.lambdify(t,ytb,"numpy")
ytt_lam= sy.lambdify(t,ytt,"numpy")
t_vals= np.linspace(0, 5, 1000)
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t_vals, ytb_lam(t_vals), linestyle="-")
axis[0,1].plot(t_vals, ytt_lam(t_vals), linestyle="-")
```

```python
b0t= sy.nsolve(ytb, t, 0)
t0t= sy.nsolve(ytt, t, 0)
print("Time for badminton ball to reach the ground: " + str(b0t) + "s;")
print("Time for tennis ball to reach the ground: " + str(t0t) + "s;")
```
