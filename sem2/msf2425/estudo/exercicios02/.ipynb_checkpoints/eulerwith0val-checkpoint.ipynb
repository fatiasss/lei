```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
g = 9.81
vm = 100/3.6
```

```python
N=10000
tfinal=2.042
yt= np.empty(N)
vt= np.empty(N)
t= np.empty(N)
yt[0]=0
vt[0]=10 #CUIDADO A COPIAR QUE V0 AQUI Ã‰ 10
t[0]=0
dt=tfinal/N
for i in range(N-1):
    a=-g
    yt[i+1]=yt[i]+vt[i]*dt
    if(vt[i]+a*dt<vm):
        vt[i+1]=vt[i]+a*dt
    else:
        vt[i+1]=vm
    t[i+1]=t[i]+dt
    if(vt[i+1]*vt[i]<0): #save when the velocity changes from negative to positive, so we know when velocity 0 was crossed
        timevel0 = ((2*i+1)/2)*dt 
        vel0index= i
    if(yt[i+1]*yt[i]<0): #save when the y changes from negative to positive, so we know when y 0 was crossed
        timey0 = ((2*i+1)/2)*dt         
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t,yt, linestyle="-")
```
