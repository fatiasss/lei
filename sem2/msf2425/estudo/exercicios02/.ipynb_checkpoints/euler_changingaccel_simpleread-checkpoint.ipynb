```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
g = 9.81
vm = 6.80
c = g/vm
```

```python
N=10000
tfinal=2
yt= np.empty(N)
vt= np.empty(N)
t= np.empty(N)
yt[0]=0
vt[0]=0
t[0]=0
dt=tfinal/N
for i in range(N-1):
    a=g- c*vt[i] #aceleração
    yt[i+1]=yt[i]+vt[i]*dt #y
    if(vt[i]+a*dt<vm): #velocidade se não chegar à terminal
        vt[i+1]=vt[i]+a*dt
    else:
        vt[i+1]=vm #velocidade set na terminal
    t[i+1]=t[i]+dt
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t,yt, linestyle="-")
axis[0,1].plot(t, vt, linestyle="-")
```
