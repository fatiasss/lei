```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
v0=200/3.6
vmax=6.8
g=9.81
```

```python
N=10000
tfinal=5
yt= np.empty(N)
vt= np.empty(N)
at=np.empty(N)
t= np.empty(N)
yt[0]=0
vt[0]=v0
t[0]=0
dt=tfinal/N
```

```python
def euler(N, yt, vt, t, dt):
    for i in range(N-1):
        at[i]=g -(g/vmax**2)*abs(vt[i])*vt[i]
        yt[i+1]=yt[i]+vt[i]*dt
        vt[i+1]=vt[i]+at[i]*dt
        t[i+1]=t[i]+dt
        if(vt[i+1]<(v0*0.5) and vt[i]>(v0*0.5)):
            print("condition")
            v50t = ((2*i+1)/2)*dt
        if(yt[i+1]>4 and yt[i]<4):
            t4m = ((2*i+1)/2)*dt
    return t4m, v50t
t4m, v50t = euler(N, yt, vt, t, dt)
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t, yt, linestyle="-")
axis[0,1].plot(t, vt, linestyle="-")
axis[1,0].plot(t, at, linestyle="-")
```
