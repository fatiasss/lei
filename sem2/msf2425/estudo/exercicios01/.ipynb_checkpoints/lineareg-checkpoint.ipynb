```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math
x = np.array([222.0, 207.5, 194.0, 171.5, 153.0, 133.0, 113.0, 92.0])
y = np.array([2.3, 2.2, 2.0, 1.8, 1.6, 1.4, 1.2, 1.0])
```

```python
def linreg(x,y):
    N = len(x)
    #sums of all elements, and sum of x**2 and y**2
    sumx = np.sum(x)
    sumy = np.sum(y)
    sumxy = np.sum(x*y)
    sumxx = np.sum(np.square(x))
    sumyy = np.sum(np.square(y))
    #formulas
    m = ((N*sumxy)-(sumx*sumy))/((N*sumxx)-sumx**2)
    b = (sumxx*sumy-sumx*sumxy)/(N*sumxx-sumx**2)
    r2 = ((N*sumxy - sumx*sumy)**2)/((N*sumxx-sumx**2)*(N*sumyy-sumy**2))
    deltam = abs(m)*math.sqrt(((1/(r2**2))-1)/(N-2))
    deltab = deltam * math.sqrt((sumxx)/N)
    return m, b, r2, deltam, deltab
    
    
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].scatter(x, y)
```

```python
m, b, r2, deltam, deltab = linreg(x, y)
m, b, r2, deltam, deltab
```

```python
axis[0,1].plot(x, m*x+b, linestyle="-")
figure
```
