```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
vmaxs=60
vmaxp=5
g=9.81
```

```python
N=100000  
tfinal=20
yt= np.zeros(N)
vt= np.zeros(N)
t= np.zeros(N)
yt[0]=800
vt[0]=0
t[0]=0
dt=tfinal/N
finalt=0
```

```python
def euler(N, yt, vt, t, dt, vmax):
    for i in range(N-1):
        a=-g -(g/vmax**2)*abs(vt[i])*vt[i]
        yt[i+1]=yt[i]+vt[i]*dt
        vt[i+1]=vt[i]+a*dt
        if abs(vt[i+1]) > vmax:
            vt[i+1]=np.sign(vt[i+1])*vmax
        t[i+1]=t[i]+dt
        if(yt[i+1]<0 and yt[i]>0):
            finalt = ((2*i+1)/2)*dt
    return finalt
finalt= euler(N, yt, vt, t, dt, vmaxs)
finalv = vt[int(finalt/dt)]
```

```python
print("O tempo para chegar ao solo sem paraquedas é: " + str(finalt) + " segundos. Neste instante tem uma velocidade de " + str(finalv) + "m/s" )
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t, yt, linestyle="-")
axis[0,1].plot(t, vt, linestyle="-")
```
