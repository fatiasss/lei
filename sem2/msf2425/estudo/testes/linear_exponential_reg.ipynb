```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math
t = np.array([0, 1, 2, 3, 4, 5, 6, 7])
n = np.array([11, 20, 33, 54, 83, 134, 244, 425])
```

```python
def linreg(x,y):
    N = len(x)
    #sums of all elements, and sum of x**2 and y**2
    sumx = np.sum(x)
    sumy = np.sum(y)
    sumxy = np.sum(x*y)
    sumxx = np.sum(np.square(x))
    sumyy = np.sum(np.square(y))
    #formulas
    m = ((N*sumxy)-(sumx*sumy))/((N*sumxx)-sumx**2)
    b = (sumxx*sumy-sumx*sumxy)/(N*sumxx-sumx**2)
    r2 = ((N*sumxy - sumx*sumy)**2)/((N*sumxx-sumx**2)*(N*sumyy-sumy**2))
    deltam = abs(m)*math.sqrt(((1/(r2**2))-1)/(N-2))
    deltab = deltam * math.sqrt((sumxx)/N)
    return m, b, r2, deltam, deltab
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].scatter(t, n)
```

```python
m, b, r2, deltam, deltab = linreg(t, n)
m, b, r2, deltam, deltab
#o coeficiente de determinação r**2 = 0.785 , pelo que o ajuste é mau. Analizando o gráfico, que curva, a relação não parece linear mas sim de potência.
```

```python
axis[0,0].plot(t, m*t+b, linestyle="-")
figure
```

```python
axis[0,1].semilogy(t, n, marker="o", linestyle="-")
figure
```

```python
m_log, b_log, r2_log, deltam_log, deltab_log = linreg(t, np.log(n))
m_log, b_log, r2_log, deltam_log, deltab_log
#o declive desta resta é  0.509, os seus erros associados são +/- 0.012 e +/- 0.052 (a sua soma para a equação em geral = 0.064) 
#e o seu coeficiente de determinação é 0.998
```

```python
axis[1,0].scatter(t, n)
axis[1,0].plot(t, np.exp(b_log)*np.exp(m_log * t), linestyle="-")
figure
#Podemos concluir que a relação é exponencial.
```
