```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math
t = np.array([0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5, 9.5, 10.5])
s = np.array([0.1, 1.4, 1.7, 6.5, 7.7, 10.4, 19.5, 26.1, 26.5, 45.9, 52.5])
```

```python
def linreg(x,y):
    N = len(x)
    #sums of all elements, and sum of x**2 and y**2
    sumx = np.sum(x)
    sumy = np.sum(y)
    sumxy = np.sum(x*y)
    sumxx = np.sum(np.square(x))
    sumyy = np.sum(np.square(y))
    #formulas
    m = ((N*sumxy)-(sumx*sumy))/((N*sumxx)-sumx**2)
    b = (sumxx*sumy-sumx*sumxy)/(N*sumxx-sumx**2)
    r2 = ((N*sumxy - sumx*sumy)**2)/((N*sumxx-sumx**2)*(N*sumyy-sumy**2))
    deltam = abs(m)*math.sqrt(((1/(r2**2))-1)/(N-2))
    deltab = deltam * math.sqrt((sumxx)/N)
    return m, b, r2, deltam, deltab
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].scatter(t, s)
```

```python
m, b, r2, deltam, deltab = linreg(t, s)
m, b, r2, deltam, deltab
#O declive é de 5.139999999999998 +/ 0.8792887992828534 , a ordenada na origem é de -10.242727272727258 +/- 5.5784620550746205 e o coeficiente de determi.
#nação é de 0.8896790414268826
```

```python
axis[0,0].plot(t, m*t+b, linestyle="-")
figure
```

```python
axis[0,1].scatter(np.log(t), np.log(s))
figure
```

```python
m_log, b_log, r2_log, deltam_log, deltab_log = linreg(np.log(t), np.log(s))
m_log, b_log, r2_log, deltam_log, deltab_log
##O declive é de 2.0114250527149666 +/ 0.11865818872736739 , a ordenada na origem é de -0.863886565978593 +/- 0.19907755141700406 e o coeficiente de determi.
#nação é de 0.9846982165348632
```

```python
axis[0,1].plot(np.log(t), m_log*np.log(t)+b_log)
axis[1,0].scatter(t, s)
axis[1,0].plot(t, np.exp(b_log)*t**m_log)
figure 
```
