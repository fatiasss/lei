```python
import sympy as sy
import matplotlib.pyplot as plt
import numpy as np
x,y,m,b = sy.symbols("x, y, m, b")
```

```python
y = m*x**2 + b
y2 = y.subs([(m,20),(b,0.0)])
y_em_1 = y2.evalf(subs={x:1})
y_lam = sy.lambdify(x,y2,"numpy")
x_vals = np.linspace(-2, 0, 100)
plot = plt.subplot(x_vals, y_lam(x_vals), linestyle="-")
plot
```

```python
yxdiff = sy.lambdify(x, sy.diff(y2,x), "numpy")
diffplot = plt.plot(x_vals, yxdiff(x_vals), linestyle="-")
diffplot
```

```python
yxintg = sy.lambdify(x, sy.integrate(y2, x), "numpy")
yxintplot = plt.plot(x_vals, yxintg(x_vals), linestyle="-")
yxintplot
```
