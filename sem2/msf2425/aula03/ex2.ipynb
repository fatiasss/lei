```python
import sympy as sy
import matplotlib.pyplot as plt
import numpy as np
figure, axis = plt.subplots(2, 2)
t, xt = sy.symbols("t, xt")
a = 3
xt = (a/2)*t**2
xtfunc = sy.lambdify(t, xt, "numpy")
t_vals = np.linspace(0, 5, 100)
axis[0,0].plot(t_vals, xtfunc(t_vals), linestyle="-")
yt= sy.lambdify(t, sy.integrate(a, t), "numpy")
axis[0,1].plot(t_vals, yt(t_vals), linestyle="-")
xt2= sy.lambdify(t, sy.integrate(sy.integrate(a, t), t), "numpy")
axis[1,0].plot(t_vals, xt2(t_vals), linestyle="-")
liftoff = sy.nsolve(sy.integrate(a, t)-69.4, t, 0)
liftoff #time for liftoff
```
