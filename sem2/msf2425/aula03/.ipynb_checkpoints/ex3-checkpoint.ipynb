```python
import sympy as sy
import matplotlib.pyplot as plt
import numpy as np
import math

t, yt = sy.symbols("t, yt")
vt= 6.8
g=9.81
t_vals = np.linspace(0, 4, 100)
figure, axis = plt.subplots(2,2)
```

```python
yt = ((vt**2)/g)*sy.log(sy.cosh((g*t)/vt))
ytfunc = sy.lambdify(t, yt, "numpy")
axis[0,0].plot(t_vals, ytfunc(t_vals), linestyle="-")
figure
```

```python
vtf = sy.diff(yt, t)
vtfunc = sy.lambdify(t, vtf, "numpy")
axis[0,1].plot(t_vals, vtfunc(t_vals), linestyle="-")
figure
```

```python
at = sy.diff(vtf, t)
atfunc = sy.lambdify(t, at, "numpy")
axis[1,0].plot(t_vals, atfunc(t_vals), linestyle="-")
figure
```

```python
atformula = g - (g/vt**2)*vtf*abs(vtf)
atformulafunc = sy.lambdify(t, atformula, "numpy")
axis[1,1].plot(t_vals, atformulafunc(t_vals), linestyle="-")
figure
```

```python
sy.nsolve(yt-20, t, 0)
```

```python
sy.nsolve(20-(9.81/2)*t**2, t, 0)
```
