```python
#João Farias, nmec. 124964
import numpy as np
import matplotlib.pyplot as plt
import math
k= 2
k1 = 0.5
m = 1
xaeq = 0  #Considerando os pontos de equilíbrio a origem.
xbeq = 0
```

```python
N=100000
tfinal=40
va0= 0.2
vb0= -0.3

xta= np.empty(N)
xta[0]=xaeq 
xtb= np.empty(N)
xtb[0]=xbeq

vta= np.empty(N)
vta[0]=va0
vtb= np.empty(N)
vtb[0]=vb0

t= np.empty(N)
t[0] = 0
dt=tfinal/N
```

```python
def eulercromer():
    for i in range(N-1):
        axa = (-k*(xta[i] -xaeq) - k1 * ((xta[i] -xaeq) - (xtb[i] - xbeq)))/m
        axb = (-k*(xtb[i] - xbeq) + k1 * ((xta[i] -xaeq) - (xtb[i] - xbeq)))/m

                
        vta[i+1]=vta[i]+axa*dt
        vtb[i+1]=vtb[i]+axb*dt
        
        
        xta[i+1]=xta[i]+vta[i+1]*dt
        xtb[i+1]=xtb[i]+vtb[i+1]*dt

        t[i+1] = t[i] + dt
        
    return
eulercromer()
```

```python
offset = 0.5 #Alteração apenas visual para melhor compreensâo do movimento

plt.plot(t, xta, linestyle="-")
plt.plot(t, xtb + offset, linestyle="-")
plt.grid(True)
plt.title("Caso i")
plt.show()
```

```python
#b) Neste caso, o movimento dos corpos é periódico, com os dois corpos sempre com velocidades opostas.
```
