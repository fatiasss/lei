```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import sympy as sy
import math
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
k = 1
k1 = 1
m = 1
A = 1
n_massas = 5
l = 1 #numero do modo normal desejado
N = 100000
tfinal = 40
dt = tfinal / N
offset = 2
```

```python
x = np.zeros((n_massas, N))
v = np.zeros((n_massas, N))
t = np.zeros(N)
```

```python
for i in range(n_massas): #Posições e velocidades iniciais para o modo l
    x[i, 0] = A * math.sin((math.pi * (i + 1) * l) / (n_massas + 1))
    v[i, 0] = 0
```

```python
def euler_cromer(): #Euler cromer para cada instante e, em cada instante, para cada massa
    for j in range(N - 1):
        for i in range(n_massas):
            f = 0
            if i > 0:
                f += -k1 * (x[i, j] - x[i - 1, j])
            if i < n_massas - 1:
                f += -k1 * (x[i, j] - x[i + 1, j])
            f += -k * x[i, j]
            a = f / m
            v[i, j + 1] = v[i, j] + a * dt
            x[i, j + 1] = x[i, j] + v[i, j + 1] * dt
        t[j + 1] = t[j] + dt
euler_cromer()
```

```python
for i in range(n_massas):
    plt.plot(t, x[i] + offset * i, label=("Massa " + str(i+1)))
plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (com offset) (m)')
plt.title('Cadeia de ' + str(n_massas) + " Osciladores - Euler-Cromer")
```
