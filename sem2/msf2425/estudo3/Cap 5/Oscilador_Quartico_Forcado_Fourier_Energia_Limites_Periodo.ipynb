```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import sympy as sy
import math
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from scipy.signal import find_peaks
m= 1
k = 1
b=0.05
alpha = 1
F0 = 7.5
wf = 1
x0= -3
v0x= -3
```

```python
N=100000
tfinal=200
xt= np.empty(N)
xt[0]=x0
vxt= np.empty(N)
vxt[0]=v0x
t = np.empty(N)
t[0] = 0

em = np.empty(N)
ep0 = 0.5 * k * (xt[0]**2)*(1+  alpha*(xt[0]**2))
ecin0 = 0.5 * m * vxt[0]**2
em[0] = ep0 + ecin0

maximum = []
minimum = []

period_list = []
amp_list = []

dt=tfinal/N
```

```python
def maxminv(x0,x1,x2,y0,y1,y2):
 # Máximo ou mínimo usando o polinómio de Lagrange
 # Dados (input): (x0,y0), (x1,y1) e (x2,y2)
 # Resultados (output): xm, ymax
 xab=x0-x1
 xac=x0-x2
 xbc=x1-x2
 a=y0/(xab*xac)
 b=-y1/(xab*xbc)
 c=y2/(xac*xbc)
 xmla=(b+c)*x0+(a+c)*x1+(a+b)*x2
 xm=0.5*xmla/(a+b+c)
 xta=xm-x0
 xtb=xm-x1
 xtc=xm-x2
 ymax=a*xtb*xtc+b*xta*xtc+c*xta*xtb
 return xm, ymax
```

```python
def eulercromer():
    for i in range(N-1):
        ax=-(k*xt[i]*(1+2*alpha*xt[i]**2) + b * vxt[i] - F0*math.cos(wf*t[i]))/m
        
        vxt[i+1]=vxt[i]+ax*dt
        xt[i+1]=xt[i]+vxt[i+1]*dt
        
        t[i+1]= t[i] + dt
        
        ep = 0.5 * k * (xt[i+1])**2
        ecin = 0.5 * m * vxt[i+1]**2
        em[i+1] = ep + ecin

        if (xt[i-1] < xt[i] > xt[i+1]) and i>=100/dt and xt[i]>1.5: #para estar no regime estacionario
            t0, t1, t2 = (i-1)*dt, i*dt, (i+1)*dt                   #and xt[i]>1.5 para remover os limites intermédios
            y0, y1, y2 = xt[i-1], xt[i], xt[i+1]
            xm, xmax = maxminv(t0, t1, t2, y0, y1, y2)
            maximum.append([xm, xmax])
        if (xt[i-1]> xt[i] < xt[i+1]) and i>=100/dt and xt[i]<0: 
            t0, t1, t2 = (i-1)*dt, i*dt, (i+1)*dt
            y0, y1, y2 = xt[i-1], xt[i], xt[i+1]
            xm, xmin = maxminv(t0, t1, t2, y0, y1, y2)
            minimum.append([xm, xmin])
        
    return
eulercromer()
```

```python
def calculate_period():
    for i in range(1, len(maximum)):
        period_list.append(maximum[i][0]-maximum[i-1][0])
    estimated_period= (sum(period_list)/len(period_list))
    return estimated_period
def calculate_amplitude():
    avg_max = sum(x[1] for x in maximum) / len(maximum)
    avg_min = sum(x[1] for x in minimum) / len(minimum)
    return abs(avg_max - avg_min) / 2, avg_max, avg_min

T = calculate_period()
Amp, avg_max, avg_min = calculate_amplitude()
print(Amp, T, avg_max, avg_min)
```

```python
 def abfourier(tp,xp,it0,it1,nf):
#
# cálculo dos coeficientes de Fourier a_nf e b_nf
#       a_nf = 2/T integral ( xp cos( nf w) ) dt   entre tp(it0) e tp(it1)
#       b_nf = 2/T integral ( xp sin( nf w) ) dt   entre tp(it0) e tp(it1)    
# integracao numerica pela aproximação trapezoidal
# input: matrizes tempo tp   (abcissas)
#                 posição xp (ordenadas) 
#       indices inicial it0
#               final   it1  (ao fim de um período)   
#       nf índice de Fourier
# output: af_bf e bf_nf  
# 
    dt=tp[1]-tp[0]
    per=tp[it1]-tp[it0]
    ome=2*np.pi/per

    s1=xp[it0]*np.cos(nf*ome*tp[it0])
    s2=xp[it1]*np.cos(nf*ome*tp[it1])
    st=xp[it0+1:it1]*np.cos(nf*ome*tp[it0+1:it1])
    soma=np.sum(st)
    
    q1=xp[it0]*np.sin(nf*ome*tp[it0])
    q2=xp[it1]*np.sin(nf*ome*tp[it1])
    qt=xp[it0+1:it1]*np.sin(nf*ome*tp[it0+1:it1])
    somq=np.sum(qt)
    
    intega=((s1+s2)/2+soma)*dt
    af=2/per*intega
    integq=((q1+q2)/2+somq)*dt
    bf=2/per*integq
    return (af,bf)
```

```python
fourier_coef = [] 
for i in range(1, 30):
    fourier_coef.append(abfourier(t, xt, 0, int(150/dt), i)) ## Em vez de N-1, 150/dt para o regime estacionario
```

```python
T = t[int(150/dt)] - t[0]  # período ## Em vez de N-1, 150/dt para o regime estacionario
omega = 2 * np.pi / T
n_vals = np.arange(1, 30)
omega_n = n_vals * omega

def separar_coeficientes(lista):
    a_n = np.array([par[0] for par in lista])
    b_n = np.array([par[1] for par in lista])
    return a_n, b_n
```

```python
an, bn = separar_coeficientes(fourier_coef)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 6))
fig.suptitle("Coeficientes de Fourier $a_n$ e $b_n$ vs $\\omega_n$", fontsize=16)

an, bn = separar_coeficientes(fourier_coef)
axs[0].stem(omega_n, an, basefmt=" ")
axs[0].set_title("Caso i) Coeficientes $a_n$")
axs[1].stem(omega_n, bn, basefmt=" ")
axs[1].set_title("Caso i) Coeficientes $b_n$")

for ax in axs.flat:
    ax.set_xlabel("$\\omega_n$ [rad/s]")
    ax.set_ylabel("Coeficiente")
    ax.grid(True)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
