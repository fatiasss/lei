```python
import numpy as np
import matplotlib.pyplot as plt
orders = np.array([[45, 90, 45, 45, 90, 45], [3, 2, 3, 2, 3, 1.5857864376269055]])
theta=0
X=0
Y=0
for i in range(len(orders[0])):
    theta+=np.radians(orders[0][i])
    u=(np.cos(theta)*orders[1][i], np.sin(theta)*orders[1][i])
    print(u)
    Xprev = X
    Yprev = Y
    X+=u[0]
    Y+=u[1]
    plt.arrow(Xprev, Yprev, X- Xprev, Y- Yprev, color='r',width=0.1,length_includes_head=True)
plt.axis('equal')
plt.show()
```
