```python
import numpy as np
import matplotlib.pyplot as plt
import math
vmax=100/3.6
g=9.81
m=0.45
r=0.11
A= math.pi*r**2
p=1.225
w= [0, 390, 0]
```

```python
N=10000
tfinal=5
xt= np.empty(N)
yt= np.empty(N)
zt= np.empty(N)
xt[0]=0
yt[0]=0
zt[0]=23.8
vxt= np.empty(N)
vyt= np.empty(N)
vzt= np.empty(N)
vxt[0]=25
vyt[0]=5
vzt[0]=-50
dt=tfinal/N
dragcoef=(g/vmax**2)
magcoef= (0.5*A*p*r*w[1])/m
```

```python
def euler():
    for i in range(N-1):
        vabs = np.sqrt(vxt[i]**2 + vyt[i]**2 + vzt[i]**2)
        ax=-dragcoef*vabs*vxt[i] + magcoef*vzt[i]
        ay=-g -dragcoef*vabs*vyt[i]
        az=-dragcoef*vabs*vzt[i] - magcoef*vxt[i]
        
        xt[i+1]=xt[i]+vxt[i]*dt
        yt[i+1]=yt[i]+vyt[i]*dt
        zt[i+1]=zt[i]+vzt[i]*dt
        
        vxt[i+1]=vxt[i]+ax*dt
        vyt[i+1]=vyt[i]+ay*dt
        vzt[i+1]=vzt[i]+az*dt
        
    return
euler()
```

```python
plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
goalx = [0,0,0,0]
goaly = [0,2.4,2.4,0]
goalz = [-3.66,-3.66,3.66,3.66]
ax.plot3D(goalx,goalz,goaly,"k")
ax.plot3D(xt[xt >= 0], -zt[xt >= 0], yt[xt >= 0], "r")
ax.set_xlim3d(0, 5)
ax.set_ylim3d(
-25, 5)
ax.set_zlim3d(0, 5)
ax.set_box_aspect((2,6,2))
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_zlabel('y')
```
