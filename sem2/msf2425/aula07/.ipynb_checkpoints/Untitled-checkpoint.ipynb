```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import sympy as sy
import math
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
v0x= 0
v0y=0
x0=0
y0=0.1
g=9.81
m=1
```

```python
N=1000
tfinal=3.3
xt= np.empty(N)
yt= np.empty(N)
xt[0]=x0
yt[0]=y0
vxt= np.empty(N)
vxt[0]=v0x
vyt= np.empty(N)
vyt[0]=-0.05
dt=tfinal/N
epi = m*g*yt[0]
ep = np.empty(N-1)
ep[0] = epi
ec = np.empty(N-1)
ec[0]=0
predictedvf = math.sqrt((epi*2)/m)
t=np.empty(N-1)
```

```python
def euler():
    for i in range(N-1):
        xt[i+1]=xt[i]+vxt[i]*dt

        

        if xt[i+1]<2:
            yt[i+1]=(0.1-0.05*xt[i+1])
            ax=g*0.05
        else:
            yt[i+1]=0
            ax=0
        
        vxt[i+1]=vxt[i]+ax*dt
        vyt[i+1]=yt[i+1]+0.001 - (yt[i]) * vxt[i+1]/0.001
        
        if(xt[i+1]>2.5 and xt[i]<2.5):
            tf = ((2*i+1)/2)*dt
            vtf = vxt[i+1]
            print(tf, vtf)
        if i!=N-1:
            ep[i] = m*g*yt[i]
            ec[i] = 0.5*m*(math.sqrt(vxt[i]**2+vyt[i]**2))
            t[i] = i*dt
        
    return vtf
vtf = euler()
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(xt, yt, linestyle="-")
axis[0,1].plot(t, ep, linestyle="-")
axis[0,1].plot(t, ec, linestyle="-")
axis[0,1].axhline(y=1, color='r', linestyle='--')
```

```python
print("Predicted final velocity =" + str(predictedvf) + "\nModel final velocity =" + str(vtf))
```
