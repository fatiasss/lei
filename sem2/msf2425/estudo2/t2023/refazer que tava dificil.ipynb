```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math
g=9.81
P=40*1000
cres= 0.25
u=0.04
par=1.225
m=2000
A=2
```

```python
N=100000
tfinal=150
xt= np.zeros(N)
vxt= np.zeros(N)
w = np.zeros(N)
t= np.zeros(N)
xt[0]=0
yt[0]=0
vxt[0]=1
t[0]=0
w[0]=0
dt=tfinal/N
goalindex=0
goaltime=0
```

```python
def euler():
    for i in range(N-1):
        theta = math.radians(5)
        fx= ((P/(vxt[i])) - m*g*math.sin(theta) - u*m*g*math.cos(theta) - 0.5*cres*A*par*(vxt[i]**2))
        ax=fx/m
        xt[i+1]=xt[i]+vxt[i]*dt
        
        vxt[i+1]=vxt[i]+ax*dt
        
        t[i+1]=t[i]+dt
        w[i+1]= w[i] + P*dt

        if(xt[i+1]>2000 and xt[i]<2000):
            goalindex=round((2*i+1)/2)
            goaltime= goalindex*dt
    return goalindex, goaltime
goalindex, goaltime = euler()
finalvs= vxt[-1]
goaltimesave = goaltime
```

```python
figure, axis= plt.subplots(2,2)
axis[0,0].plot(t,xt, linestyle="-")
axis[0,1].plot(xt,vxt, linestyle="-")
axis[1,0].plot(t,w, linestyle="-")
```

```python
print("Atinge os 2km aos: " + str(float(goaltime)) + " segundos")
print("Realiza um trabalho de " + str(float(w[goalindex])) + " J")
```
