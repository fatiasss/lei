```python
import numpy as np
import matplotlib.pyplot as plt
import math
vmax=20
g=9.81
m=0.057
r=0.034
A= math.pi*r**2
p=1.225
w= [0, 0, 0]
```

```python
N=10000
tfinal=1
xt= np.empty(N)
yt= np.empty(N)
zt= np.empty(N)
xt[0]=0
yt[0]=3
zt[0]=0
vxt= np.empty(N)
vyt= np.empty(N)
vzt= np.empty(N)
t= np.zeros(N)
vxt[0]=30
vyt[0]=0
vzt[0]=0
dt=tfinal/N
dragcoef=(g/vmax**2)
magcoef= (0.5*A*p*r)/m
```

```python
def euler():
    for i in range(N-1):
        wvcross = np.cross(w, [vxt[i], vyt[i], vzt[i]]) #crossproduct of w and v
        vabs = np.sqrt(vxt[i]**2 + vyt[i]**2 + vzt[i]**2)
        
        ax=-dragcoef*vabs*vxt[i] + magcoef*wvcross[0]
        ay=-g -dragcoef*vabs*vyt[i] + magcoef*wvcross[1]
        az=-dragcoef*vabs*vzt[i] + magcoef*wvcross[2]
        
        xt[i+1]=xt[i]+vxt[i]*dt
        yt[i+1]=yt[i]+vyt[i]*dt
        zt[i+1]=zt[i]+vzt[i]*dt
        
        vxt[i+1]=vxt[i]+ax*dt
        vyt[i+1]=vyt[i]+ay*dt
        vzt[i+1]=vzt[i]+az*dt

        t[i+1]=t[i]+ dt
        if(yt[i+1]*yt[i]<0): #save when the y changes from positive to negative, so we know when y 0 was crossed
            timey0 = ((2*i+1)/2)*dt
            y0index = i
        if(xt[i+1]>12 and xt[i]<12): #save when the y changes from positive to negative, so we know when y 0 was crossed
            timenet = ((2*i+1)/2)*dt
            netindex = i
    return timey0, y0index, timenet, netindex
timey0, y0index, timenet, netindex = euler()
```

```python
figure, axis = plt.subplots(2, 2)

axis[0, 0].plot(xt, yt, linestyle="-")
```

```python
print("Quando a bola passa pela rede, o seu y é:" + str(float(yt[netindex]))) #logo passa por cima da rede
```

```python
print("Quando a bola aterra, o seu x é: " + str(float(xt[y0index]))) #Pelo que o serviço não é válido, já que 19.687 > 18.4 (m)
```
