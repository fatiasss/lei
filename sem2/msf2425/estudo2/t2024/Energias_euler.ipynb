```python
import sympy as sy
import matplotlib.pyplot as plt
import numpy as np
import math

k=2
m=0.5
t, ep, xt = sy.symbols("t, ep, xt")
x_vals = np.linspace(-1, 1, 100)
figure, axis = plt.subplots(1,2)
```

```python
ep= k*(((xt-0.5)**2)*(xt+0.5)**2)
epfunc = sy.lambdify(xt, ep, "numpy")
axis[0].plot(x_vals, epfunc(x_vals), linestyle="-")
figure
```

```python
N=100000
tfinal=5
xt= np.zeros(N)
vxt= np.zeros(N)
t= np.zeros(N)
ecin= np.zeros(N)
epe = np.zeros(N)
et = np.zeros(N)
xt[0]=1
vxt[0]=0
t[0]=0
ecin[0]= 0
epe[0]= 0
et[0]= 0
dt=tfinal/N
```

```python
def euler():
    for i in range(N-1):
        fx= -4*k*xt[i]**3 + k*xt[i]
        ax=fx/m
        xt[i+1]=xt[i]+vxt[i]*dt
        
        vxt[i+1]=vxt[i]+ax*dt
        
        t[i+1]=t[i]+dt
        ecin[i] = 0.5*m*vxt[i]**2
        epe[i]= k*(((xt[i]-0.5)**2)*(xt[i]+0.5)**2)
        et[i] = ecin[i] + epe[i]

        if(xt[i+1]>0.5 and xt[i]<0.5):
            dvel=vxt[round((2*i+1)/2)]
    return dvel
dvel = euler()
```

```python
axis[1].plot(t, xt, linestyle="-")
figure
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(t[0:-2], ecin[0:-2], linestyle="-")
ax[0,0].plot(t[0:-2], epe[0:-2], linestyle="-")
ax[0,0].plot(t[0:-2], et[0:-2], linestyle="-")
ax[0,1].plot(t[0:-2], et[0:-2], linestyle="-")
```

```python
print("A velocidade em x=0.5 : " + str(float(dvel)))
```
