```python
import numpy as np
import matplotlib.pyplot as plt
import math
v0= 100/3.6
g=9.81
m=0.45
r=0.11
A= math.pi*r**2
p=1.225
w= [0, 0, 0]
D= 0.0127
```

```python
N=10000
tfinal=1
xt= np.empty(N)
yt= np.empty(N)
zt= np.empty(N)
xt[0]=0
yt[0]=0
zt[0]=0
vxt= np.empty(N)
vyt= np.empty(N)
vzt= np.empty(N)
t= np.zeros(N)
vxt[0]=v0*math.cos(math.radians(16))
vyt[0]=v0*math.sin(math.radians(16))
vzt[0]=0
dt=tfinal/N
dragcoef=m*D
magcoef= (0.5*A*p*r)/m
```

```python
def euler():
    for i in range(N-1):
        wvcross = np.cross(w, [vxt[i], vyt[i], vzt[i]]) #crossproduct of w and v
        vabs = np.sqrt(vxt[i]**2 + vyt[i]**2 + vzt[i]**2)
        
        ax=-dragcoef*vabs*vxt[i] + magcoef*wvcross[0]
        ay=-g -dragcoef*vabs*vyt[i] + magcoef*wvcross[1]
        az=-dragcoef*vabs*vzt[i] + magcoef*wvcross[2]
        
        xt[i+1]=xt[i]+vxt[i]*dt
        yt[i+1]=yt[i]+vyt[i]*dt
        zt[i+1]=zt[i]+vzt[i]*dt
        
        vxt[i+1]=vxt[i]+ax*dt
        vyt[i+1]=vyt[i]+ay*dt
        vzt[i+1]=vzt[i]+az*dt

        t[i+1]=t[i]+ dt
        if(xt[i+1]>20 and xt[i]<20): 
            timegoal = ((2*i+1)/2)*dt
            goalindex = i
    return timegoal, goalindex
timegoal, goalindex = euler()
```

```python
figure, axis = plt.subplots(2, 2)

axis[0, 0].plot(xt, yt, linestyle="-")
```

```python
print("Quando a bola passa pela baliza, o seu y é:" + str(float(yt[goalindex]))) #logo passa por cima da baliza
print("Quando a bola passa pela baliza, o seu z é:" + str(float(zt[goalindex]))) #concluimos que não entra na baliza
```
