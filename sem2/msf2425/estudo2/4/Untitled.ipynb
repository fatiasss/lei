```python
import sympy as sy
import matplotlib.pyplot as plt
import numpy as np
import math

t, xt, ep= sy.symbols("t, xt, ep")
k=1
xeq=2
x_vals = np.linspace(-5, 5, 1000)
figure, axis = plt.subplots(2,2)
```

```python
ep = 0.5*k*(abs(xt)-xeq)**2
epfunc = sy.lambdify(xt, ep, "numpy")
axis[0,0].plot(x_vals, epfunc(x_vals), linestyle="-")
figure
```
