```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import sympy as sy
import math
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
v0x= 0
x0=4
k=1
m=1
A=4 #amplitude
phi= 0 #phase
```

```python
N=10000
tfinal=10
xt= np.empty(N)
xt[0]=x0
vxt= np.empty(N)
vxt[0]=v0x
t = np.empty(N)
t[0] = 0
dt=tfinal/N
et= np.empty(N)
```

```python
def eulercromer():
    for i in range(N-1):
        ax=-(k*xt[i])/m
        vxt[i+1]=vxt[i]+ax*dt
        xt[i+1]=xt[i]+vxt[i+1]*dt
        t[i+1]= t[i] + dt
        et[i]= 0.5*k*(xt[i]**2) + 0.5*m*(vxt[i]**2)
    return
eulercromer()
```

```python
figure, axis = plt.subplots(2,2)
axis[0,0].plot(t, xt, linestyle="-")
axis[0,1].plot(t[0:-2], et[0:-2], linestyle="-")
```
