```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math
g=9.81
```

```python
N=10000
tfinal=1
xt= np.zeros(N)
yt= np.zeros(N)
vxt= np.zeros(N)
vyt= np.zeros(N)
t= np.zeros(N)
xt[0]=0
yt[0]=0
vxt[0]=math.cos(math.radians(10))*(100/3.6)
vyt[0]=math.sin(math.radians(10))*(100/3.6)
t[0]=0
dt=tfinal/N
vt=100/3.6
timevel0= 0
vel0index= 0
timey0= 0
```

```python
def euler():
    for i in range(N-1):
        vabs =math.sqrt((vxt[i]**2)+(vyt[i]**2))
        ax= -(g/(vt**2))*vabs*vxt[i]
        ay= -g -(g/(vt**2))*vabs*vyt[i]
        
        xt[i+1]=xt[i]+vxt[i]*dt
        yt[i+1]=yt[i]+vyt[i]*dt
        
        vxt[i+1]=vxt[i]+ax*dt
        vyt[i+1]=vyt[i]+ay*dt
        
        t[i+1]=t[i]+dt
        if(vyt[i+1]*vyt[i]<0): #save when the vertical velocity changes from positive to negative, so we know when velocity 0 was crossed
            timevel0 = ((2*i+1)/2)*dt 
            vel0index= i
        if(yt[i+1]*yt[i]<0): #save when the y changes from positive to negative, so we know when y 0 was crossed
            timey0 = ((2*i+1)/2)*dt
            y0index = i
    return timevel0, vel0index, timey0, y0index
timevel0, vel0index, timey0, y0index = euler()
```

```python
figure, axis= plt.subplots(2,2)
axis[0,0].plot(t,yt, linestyle="-")
```

```python
print("A bola atinge a sua altura máxima aos {:.2f} segundos, atingindo os {:.2f} metros.".format(timevel0, yt[vel0index]))
print("A bola chega ao chão aos {:.2f} segundos, atingindo os {:.2f} metros.".format(timey0, xt[y0index]))
```
