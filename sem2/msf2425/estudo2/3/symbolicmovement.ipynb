```python
import sympy as sy
import matplotlib.pyplot as plt
import numpy as np
import math

t, yt, xt = sy.symbols("t, yt, xt")
g=9.81
t_vals = np.linspace(0, 1, 100)
figure, axis = plt.subplots(2,2)
```

```python
yt = sy.sin(math.radians(10))*(100/3.6)*t - (g/2)*t**2
ytfunc = sy.lambdify(t, yt, "numpy")
axis[0,0].plot(t_vals, ytfunc(t_vals), linestyle="-")
figure
```

```python
xt= sy.cos(math.radians(10))*(100/3.6)*t
xtfunc = sy.lambdify(t, xt, "numpy")
axis[0,1].plot(t_vals, xtfunc(t_vals), linestyle="-")
figure
```

```python
axis[1,0].plot(xtfunc(t_vals), ytfunc(t_vals), linestyle="-")
figure
```

```python
vy= sy.diff(yt,t)
vyfunc = sy.lambdify(t, vy, "numpy")
axis[1,1].plot(t_vals, vyfunc(t_vals), linestyle="-")
figure
```

```python
dp = sy.nsolve(vy, t, 0) 
```

```python
ytfunc(dp) #altura maxima
```

```python
tf = sy.nsolve(yt, t, 0.5)  #tempo para yt= 0
xtfunc(tf) #alcance
```
