```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import sympy as sy
import math
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
L=1
g=9.81
theta0=0.0
```

```python
def starting_conditions():
    global N, tfinal, xt, yt, ang, vt, dt, maximum, period_list
    N=10000
    tfinal=10
    xt= np.empty(N)
    yt= np.empty(N)
    ang=np.empty(N)
    ang[0]=theta0
    xt[0]=L*math.sin(theta0)
    yt[0]=-L*math.cos(theta0)
    vt= np.empty(N)
    vt[0]=0
    dt=tfinal/N
    maximum= []
    period_list=[]
starting_conditions()
```

```python
def maxminv(x0,x1,x2,y0,y1,y2):
 # Máximo ou mínimo usando o polinómio de Lagrange
 # Dados (input): (x0,y0), (x1,y1) e (x2,y2)
 # Resultados (output): xm, ymax
 xab=x0-x1
 xac=x0-x2
 xbc=x1-x2
 a=y0/(xab*xac)
 b=-y1/(xab*xbc)
 c=y2/(xac*xbc)
 xmla=(b+c)*x0+(a+c)*x1+(a+b)*x2
 xm=0.5*xmla/(a+b+c)
 xta=xm-x0
 xtb=xm-x1
 xtc=xm-x2
 ymax=a*xtb*xtc+b*xta*xtc+c*xta*xtb
 return xm, ymax
```

```python
def eulercromer():
    for i in range(N-1):
        atheta=-(g/L) * math.sin(ang[i])

        vt[i+1] = vt[i]+atheta*dt

        ang[i+1] = ang[i]+vt[i+1]*dt

        xt[i+1] = math.sin(ang[i+1])*L
        yt[i+1] = -(math.cos(ang[i+1])*L)

        if yt[i-1] < yt[i] > yt[i+1]:
            t0, t1, t2 = (i-1)*dt, i*dt, (i+1)*dt
            y0, y1, y2 = yt[i-1], yt[i], yt[i+1]
            xm, ymax = maxminv(t0, t1, t2, y0, y1, y2)
            maximum.append([xm, ymax])
        
    return
eulercromer()
```

```python
def calculate_period():
    for i in range(1, len(maximum)):
        period_list.append(maximum[i][0]-maximum[i-1][0])
    estimated_period= 2*(sum(period_list)/len(period_list))
    return estimated_period
print(calculate_period())
```

```python
def run_sim():
    starting_conditions()
    eulercromer()
    return calculate_period()


def L_vals():
    vals = []
    for i in range(1, 10):
        global L
        L=i
        period = run_sim()
        vals.append([math.log(L), math.log(period)])
    vals = np.array(vals)
    return vals
vals = L_vals()
x_vals = vals[:, 0]
y_vals = vals[:, 1]
```

```python
def linreg(x,y):
    N = len(x)
    #sums of all elements, and sum of x**2 and y**2
    sumx = np.sum(x)
    sumy = np.sum(y)
    sumxy = np.sum(x*y)
    sumxx = np.sum(np.square(x))
    sumyy = np.sum(np.square(y))
    #formulas
    m = ((N*sumxy)-(sumx*sumy))/((N*sumxx)-sumx**2)
    b = (sumxx*sumy-sumx*sumxy)/(N*sumxx-sumx**2)
    r2 = ((N*sumxy - sumx*sumy)**2)/((N*sumxx-sumx**2)*(N*sumyy-sumy**2))
    deltam = abs(m)*math.sqrt(((1/(r2**2))-1)/(N-2))
    deltab = deltam * math.sqrt((sumxx)/N)
    return m, b, r2, deltam, deltab
m, b, r2, deltam, deltab = linreg(x_vals, y_vals)
```

```python
x_vals = vals[:, 0]
y_vals = vals[:, 1]

plt.plot(x_vals, y_vals, 'o', label='log(period) vs log(L)')
plt.plot(x_vals, m*x_vals+b, linestyle="-")
plt.xlabel('log(L)')
plt.ylabel('log(period)')
plt.title('Log-Log Plot of Period vs Length')
plt.legend()
plt.grid(True)
plt.show()
```
