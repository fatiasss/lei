```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
g=9.8
c= 2.5
m=50/1000
```

```python
N=10000
tfinal=20
yt= np.zeros(N)
vt= np.zeros(N)
t= np.zeros(N)
yt[0]=30/100
vt[0]=0
t[0]=0
dt=tfinal/N
```

```python
def euler(N, yt, vt, t, dt):
    for i in range(N-1):
        if(yt[i]>0):
            a=-g
        else:
            a=-g -(c*vt[i])/m
        yt[i+1]=yt[i]+vt[i]*dt
        vt[i+1]=vt[i]+a*dt
        t[i+1]=t[i]+dt
    return 
euler(N, yt, vt, t, dt)
```

```python
figure, axis = plt.subplots(2,1)
axis[0].plot(t, yt, linestyle="-")
axis[0].set_xlabel("t / s")
axis[0].set_ylabel("yt / m")
axis[1].plot(t, vt, linestyle="-")
axis[1].set_xlabel("t / s")
axis[1].set_ylabel("vt / m/s")
figure.tight_layout(pad=2.0)
```

```python
vf = vt[-1]
#A velocidade final atingida no líquido é -0.1962 m/s. Esta velocidade relaciona-se com C da seguinte forma:
# C = -(m * g)/vf, ou seja C = -P/vf ,
# sendo aqui m a massa do objeto, g a aceleração gravítica, vf a velocidade final do objeto e P a intensidade dd seu peso.
```
